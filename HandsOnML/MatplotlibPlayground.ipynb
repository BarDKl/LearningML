```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use(style = 'seaborn-v0_8-dark-palette')
fig = plt.figure()
ax = plt.axes()

x = np.linspace(start = 0,stop = 10,num = 50)
plt.scatter(x,np.sin(x))
plt.scatter(x,np.cos(x))
plt.axis('tight')
plt.show()
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = y = np.linspace(0,5,40)
X, Y = np.meshgrid(x,y)
Z = f(X,Y)
plt.contourf(X, Y, Z, 200)
plt.colorbar()
```
